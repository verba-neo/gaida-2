```python
# OPENAI API를 사용하기 -> GPT한테 질문을 하고, 답을 받겠다.

# 1. OPENAI 와 통신할 SW설치
%pip install openai
```

```python
prompt="""
농담 하나 가봐라
"""

from openai import OpenAI
client = OpenAI(api_key='<YOUR-KEY-HERE>')

response = client.responses.create(
    model='gpt-4.1-nano',
    input=prompt
)

print(response.output_text)
```
