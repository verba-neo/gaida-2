````python
# name: plus_one / in: 숫자 / out: 숫자 + 1

# 함수 정의(define)
def plus_one(num):
    return num + 1


# 함수 호출(call)
plus_one(999)
````

````python
# n번 인사하는 함수
# name: hello_ntimes

def hello_ntimes( n ):
    # for-매개변수에 들어간 값을 사용하지 않을 때 => _
    for _ in range(n):
        print('반가워')


hello_ntimes(5)  # 5번 hello를 출력
````

````python
# 3제곱 해주는 함수
# name: cube, in: 숫자 1개, out: 숫자의 세제곱

def cube(n):
    return n ** 3

cube(5)  # 27
````

````python
# 원기둥의 부피를 구하는 함수
# name: cylinder, in: 반지름/높이, out: 부피(pi * r^2 * h)
import math

def cylinder(r, h):
    volume = math.pi * r**2 * h
    return volume

print(cylinder(2, 5))  # 반지름2, 높이5
round(cylinder(5, 2), 2)  # 반올림 함수
````

````python
# 리스트 내부의 숫자들의 평균을 구해주는 함수
# name: get_avg, in: 숫자들로 만들어진 리스트, out: 평균
def get_avg(numbers):
    avg = sum(numbers) / len(numbers)
    return avg

get_avg([1, 2, 3, 4, 19, 23, 45, 88])  # 3
````

````python
# 비밀번호 체크해주는 함수 (숫자 포함, 8글자 이상)
# name: check_pw, in: 비밀번호 문자열, out: True/False

def check_pw(password):
    is_ok = False
    if len(password) >= 8:
        for char in password:
            if char in '0123456789':
                is_ok = True
    return is_ok

check_pw('abcd1234')  # True
check_pw('abcdefg')  # False
````
