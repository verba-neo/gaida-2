## 비교 연산자

### Short-Circuit Evaluation (단축 평가)
- 앞에서 값이 확실시 되면, 그 뒤는 확인하지 않는다.
- 조건 판별시 뒷부분을 확인하지 않아도 됨.

```python
a = True and True and True and False and True and False
#     ?   ->  ?   ->   ?   ->   !  [종료, 뒤에 볼 필요 없음]
b = False or False or True or False
#     ?   ->  ?   ->   !

print(a, b)
```

## 복합 연산자

| 연산자   | 내용             |
|----------|------------------|
| `a += b`  | `a = a + b`      |
| `a -= b`  | `a = a - b`      |
| `a *= b`  | `a = a * b`      |
| `a /= b`  | `a = a / b`      |

| `a **= b` | `a = a ** b`     |

```python
# bmi 계산기
# bmi = 몸무게(kg) / 키(m) 제곱 2
# weight 에 몸무게, height 에 키를 저장하고 bmi 를 계산하기
weight = 70
height = 1.75

bmi = weight / (height**2)
print(bmi)
```

```python
# 원 넓이와 둘레 계산 (pi 는 3.14 로 계산)
# 원 넓이 = pi x r^2
# 원 둘레 = 2 x pi x r

from math import pi
# pi = 3.14
r = 3

area = pi * (r**2)
perimeter = 2 * pi * r

print(area, perimeter)
```

```python
# 초를 시간, 분, 초 로 바꾸기
# 계산을 통해 h, m, s 에 시간, 분, 초로 전환

seconds = 3675  # 1시간 1분 15초

# 시간을 구하고
h = seconds // (60 * 60)
# 그만큼 뺀다
seconds = seconds - h * 3600

# 분을 구하고
m = seconds // (60)
# 그만큼 뺀다
seconds = seconds - m * 60

# 남은게 초
s = seconds

print(f'{h}시간 {m}분 {s}초')
```

```python
# 초를 시간, 분, 초 로 바꾸기
# 계산을 통해 h, m, s 에 시간, 분, 초로 전환

seconds = 3675  # 1시간 1분 15초

h = seconds // 3600
seconds %= 3600  

m = seconds // 60
seconds %= 60

s = seconds
# h, seconds = divmod(seconds, 3600)
# m, seconds = divmod(seconds, 60)
# s = seconds

print(f'{h}시간 {m}분 {s}초')
```

```python
# 세자리 수 분리하기
# 숫자를 백, 십, 일의 자리로 분리하여 변수에 저장하기

n = 472

hund = n // 100
tens = (n % 100) // 10
ones = n % 10
print(hund, tens, ones)
```
