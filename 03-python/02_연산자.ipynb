## 산술 연산자

|연산자|내용|
|---|---|
|`+`|더하기, 양수|
|`-`|빼기, 음수|
|`*`|곱하기|
|`/`|나누기|
|`//`|몫|
|`%`|나머지|
|`**`|거듭제곱|



```python
# 2의 1000제곱
2 ** 1000
```

```python
# 정수 <-> 정수 연산은 정수가 대부분이지만, /는 실수가 나온다 
x = 3 / 2
y = 4 / 2
z = 8 % 3

x, y, z
```

## 비교 연산자
|연산자|내용|
|---|---|
|`>`, `<`|초과, 미만|
|`>=`, `<=`|이상, 이하|
|`==`|값이 같음|
|`!=`|다르다|
|`in`|포함여부|
|`is`|동일하다|

```python
# 크고 작음

5 > 3  # T
4 < 2  # F
4 >= 4  # T
3 <= 1  # F
```

```python
5 == 3  # F
4 != 2  # T
'apple' == 'banana'
```

```python
# x in y -> y안에 x가 있는지

'i' in 'apple'  # False
3 in [1, 2, 3, 4]  # True
```
