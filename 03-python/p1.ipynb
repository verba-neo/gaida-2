```python
scores = [80, 75, 90, 60, 100]

# avg = sum(scores) / len(scores)

total = 0
for score in scores:
    total += score

avg = total / len(scores)

print(avg)
if avg >= 80:
    print('합격')
else:
    print('불합격')
```

```python
numbers = [3, 7, 10, 2, 9, 5]
# max(numbers)

# 우선 최대값 후보로 첫번째 숫자를 잡는다.
maximum = numbers[0]

for num in numbers:
    if num > maximum:
        maximum = num

print(maximum)
```

```python
numbers = [1, 2, 3, 4, 5, 6, 7, 8]
# 짝수만 저장해 둘 리스트 생성
evens = []

for num in numbers:
    if num % 2 == 0:
        evens.append(num)

print(evens)
```

```python
target = "철수"
names = ["철수", "영희", "민수", "지영"]

# in 연산자로 쉽게도 가능!
print(target in names)

result = '없습니다'

for name in names:
    if name == target:
        result = '있습니다'

print(result)
```

```python
inventory = {
    "apple": 10,
    "banana": 0,
    "orange": 5
}

for key in inventory:
    if inventory[key] > 0:
        print(key, '판매가능')
    else:
        print(key, '품절')    
```

```python
# 문제 7. 리스트 안 딕셔너리 평균 구하기
students = [
    {"name": "철수", "score": 80},
    {"name": "영희", "score": 90},
    {"name": "민수", "score": 70}
]

# 전체 평균 점수
total = 0
for student in students:
    total += student['score']
avg = total / len(students)
print('평균점수', avg)

# 평균 이상인 학생 출력
for student in students:
    if student['score'] >= avg:
        print(student['name'])
```

```python
# 문제 8. 등급 매기기
scores = [95, 82, 67, 74, 88]

grades = []

for score in scores:
    if score >= 90:
        grade = 'A'
    elif score >= 80:
        grade = 'B'
    elif score >= 70:
        grade = 'C'
    else:
        grade = 'D'
    
    grades.append(grade)

print(grades)
```

```python
# 문제 9. 장바구니 총 금액 계산
cart = [
    {"item": "노트북", "price": 1000000, "qty": 1},
    {"item": "마우스", "price": 20000, "qty": 2}
]

total = 0
for c in cart:
    total += c['price'] * c['qty']

print(total)
if total >= 1e6:  # 1 * 10^6
    print('고액구매자')
```

```python
# 문제 10. 비밀번호 검사기
password = "abcgfd123"

# 틀렸다고 가정
is_ok = False

# 8글자 이상
if len(password) >= 8:
    # 한 글자씩 돌면서 숫자 체크
    for char in password:
        # if char.isdigit():
        if char in '0123456789':
            is_ok = True

print(is_ok)
```
