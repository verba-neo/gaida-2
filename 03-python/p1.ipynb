```python
scores = [80, 75, 90, 60, 100]

# avg = sum(scores) / len(scores)

total = 0
for score in scores:
    total += score

avg = total / len(scores)

print(avg)
if avg >= 80:
    print('합격')
else:
    print('불합격')
```

```python
numbers = [3, 7, 10, 2, 9, 5]
# max(numbers)

# 우선 최대값 후보로 첫번째 숫자를 잡는다.
maximum = numbers[0]

for num in numbers:
    if num > maximum:
        maximum = num

print(maximum)
```

```python
numbers = [1, 2, 3, 4, 5, 6, 7, 8]
# 짝수만 저장해 둘 리스트 생성
evens = []

for num in numbers:
    if num % 2 == 0:
        evens.append(num)

print(evens)
```

```python
target = "철수"
names = ["철수", "영희", "민수", "지영"]

# in 연산자로 쉽게도 가능!
print(target in names)

result = '없습니다'

for name in names:
    if name == target:
        result = '있습니다'

print(result)
```
