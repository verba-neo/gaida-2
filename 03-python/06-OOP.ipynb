### OOP(객체지향언어) -> 실존하는 값이 주인공
- Object : 객체 : 값 (`1`, `'abc'`, `[1, 2, 3]`)
- Oriented : 지향 : 핵심, 주인공

- Class - Instance

```python
# Person 
# 키, 몸무게, 나이, 이름
# 인사하다, 공부하다

# python에서 클래스 이름은 PascalCase로 짓는다 my_good_friend / MyGoodFriend
class Person:
    # 1. 특성 (attribute)
    height = 165
    weight = 55
    age = 20
    iq = 100
    name = '인간'
    
    # 2. 행동 (method)
    def greeting(self):
        return f'안녕하세요, {self.name} 입니다. {self.age}살 입니다.'

    def study(self):
        self.iq += 5
```

```python
# Person 개념적 대상을 -> 실제 생성
p1 = Person()
p2 = Person()
```

```python
# 속성
p1.age = 30
p1.name = '김로봇'
p1.weight = 80
p1.height = 230
# 행동
p1.greeting()
```

```python
p2.age = 23
p2.name = '박사람'
p2.weight = 45
p2.height = 155
p2.greeting()
```

```python
r1 = range(100)  # 레인지 클래스의 인스턴스를 만든다
l1 = list(r1)    # 리스트 클래스의 인스턴스를 ""

a = 3 / 1

type('abcd')
```

```python
# 객체 지향 세상 -> 객체(값)들끼리 상호작용(메서드) 하는 세상
1 + 2
(2).__add__(3)

3 > 2
(3).__gt__(2)

[1, 2, 3][2]
([1, 2, 3]).__getitem__(1)
```
