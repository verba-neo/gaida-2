# 04_조건_반복

```python
# 조건 - if

n = -10

if n > 0:
    # 블럭 안은 조건이 True 일때만 실행
    print('양수 입니다')
    print('11111')

# 블럭 밖은 조건문과 상관 없음
print('22222')
```

```python
# 조건 - if-else

n = 0

if n > 0:
    print('0보다 크다.')
else:
    print('0보다 작거나 같다.')
```

```python
# 조건 - if-elif-else
n = -10

if n > 0:
    print('양수')
elif n == 0:
    print('0이다')
else:
    print('음수')
```

```python
# 불합: 0-399
# 1: 400-599
# 2: 600-749
# 3: 750-899
# 4: 900-1000
score = 980

if score >= 900:
    print('level 4')
elif score >= 750:
    print('level 3')
elif score >= 600:
    print('level 2')
elif score >= 400:
    print('leve 1')
else:
    print('ㅠㅠ')
```

```python
# 우수 학생 판별
status = {
    'score': 60,
    'absent': 4
}

# 1. 점수가 70을 넘는 학생이면 우수학생
# 2. 우수학생 중에 결석일이 5일 미만이면 최우수학생
# 3. (우수 아닌) 결석일이 5일 미만이면 성실학생
# 4. 나머지는 일반 학생
if status['score'] >= 70:
    if status['absent'] < 5:
        print('최우수학생')
    else:
        print('우수학생')
elif status ['absent'] < 5:
    print('성실학생')
else:
    print('일반학생')
```
